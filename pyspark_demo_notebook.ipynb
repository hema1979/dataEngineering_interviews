# PySpark Demo Notebook (Medallion Architecture)

This notebook demonstrates a simple end-to-end PySpark pipeline including:
- Data Ingestion (Bronze Layer)
- Data Transformation (Silver Layer)
- Aggregation (Gold Layer)
- Basic Optimization Techniques

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('PySparkDemo').getOrCreate()
spark
```

## Bronze Layer - Data Ingestion

```python
data = [
    (1, 'Alice', 'Electronics', 200, 'active'),
    (2, 'Bob', 'Clothing', 150, 'inactive'),
    (3, 'Cathy', 'Electronics', 300, 'active'),
    (4, 'David', 'Furniture', None, 'active')
]
columns = ['id', 'name', 'category', 'amount', 'status']
bronze_df = spark.createDataFrame(data, columns)
bronze_df.show()
```

## Silver Layer - Data Cleaning & Transformation

```python
from pyspark.sql.functions import col
silver_df = bronze_df.filter(col('status') == 'active')
silver_df = silver_df.fillna({'amount': 0})
silver_df = silver_df.withColumn('amount_with_tax', col('amount') * 1.1)
silver_df.show()
```

## Gold Layer - Aggregation

```python
gold_df = silver_df.groupBy('category').sum('amount_with_tax')
gold_df.show()
```
